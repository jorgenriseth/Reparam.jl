```julia
using Plots
using LinearAlgebra
```

```julia
include("src/projection.jl")
include("src/functionals.jl")
```

```julia
γ(t) = 0.9t^2 + 0.1t
φ(t) = (1 + exp(-20)) / (1 + exp(-40(t-0.5)))
τ(t) = log(10t+1) / log(11)
```

```julia
function project(γ, N)
    projector = FourierSineProjector(N)
    
    g(t) = γ(t) - t
    gproj = project(g, projector)
    γproj = t -> gproj(t) + t

    X = 0:0.001:1
    plot(γ, X)
    plot!(γproj, X)
end
```

```julia
project(γ, 10)
```

```julia
project(φ, 15)
```

```julia
project(τ, 20)
```
