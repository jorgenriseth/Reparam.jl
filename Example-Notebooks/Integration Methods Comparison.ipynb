```julia
using QuadGK
using FastGaussQuadrature
using Plots
```

```julia
function approx_integration(f, N; a=0, b=1) 
    out = (f(a) + f(b)) / 2.
    
    h = (b - a) / N
    for i in 1:N-1
        out += f(i*h)
    end
    
    return out  * h
end
```

```julia
v(t) = 3t^2 * exp(t^3)
```

```julia
plot(v, 0:0.001:1)
```

```julia
using Printf
```

```julia
@printf "%4s %20s %20s\n" "N" "Approximate" "Error"
for n in 1:1000
    I = approx_integration(v, n)
    @printf "%4d %20.15e %20.15e\n" n I abs(I - (exp(1) - 1))
end
```
